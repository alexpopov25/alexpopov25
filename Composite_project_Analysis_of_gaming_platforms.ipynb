# Сборный проект: анализ игровых платформ

В нашем распоряжении имеются исторические данные о продажах компьютерных игр, оценки пользователей и экспертов, жанры и игровые платформы (Xbox, PlayStation и другие).

**Описание данных:**
* Name — название игры  
* Platform — платформа  
* Year_of_Release — год выпуска  
* Genre — жанр игры  
* NA_sales — продажи в Северной Америке (миллионы проданных копий)  
* EU_sales — продажи в Европе (миллионы проданных копий)  
* JP_sales — продажи в Японии (миллионы проданных копий)  
* Other_sales — продажи в других странах (миллионы проданных копий)  
* Critic_Score — оценка критиков (максимум 100)  
* User_Score — оценка пользователей (максимум 10)  
* Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.  

**Цель проекта:**
Выявить определяющие успешность игры закономерности для того чтобы отдел маркетинга интернет магазина "Стримчик" смог сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

**План исследования:**

* Изучим общую информацию

* Сформулируем общий вывод и рекомендации

## Предобработка данных

### Заменa названия столбцов

```python
# приведем названия столбцов к нижнему регистру
data.columns = data.columns.str.lower()
data.columns
```

### Поиск и обработка пропусков

Перед преобразованием типов данных в столбцах целесообразно сначала найти и изучить пропущенные значения в данных, так как пропуски в данных (NaN) не позволят произвести автоматическое преобразование типа данных.

```python
# ищем пропуски
data.isna().sum()
```

Из результатов видно, что пропуски имеются в столбцах: name, year_of_release, genre, critic_score, user_score, rating.

Для наглядности, построим температурный график по количеству пропусков:

```python
pd.DataFrame(data.isna().mean().to_frame(name='Пропуски').query('Пропуски > 0')['Пропуски'])\
            .style.background_gradient('coolwarm')\
            .format({'Пропуски':'{:.0%}'})
```

Теперь посмотрим, совпадают ли пропуски в данных по critic_score, user_score, rating по строкам. Для этого проведём визуализацию данных таблицы.

```python
# построим тепловую карту, где наличие данных обозначено синим цветом, а отстутствие - красным
plt.figure(figsize=(9,9))
sns.heatmap(data.isnull(), cmap=sns.color_palette(['#000099', '#FF3030']))
plt.title('Тепловая карта пропусков')
plt.xlabel('Название столбца')
plt.ylabel('Номер строки')
plt.show()
```

Теперь мы чётче видим, что пропуски в таблице в основном в оценках игр и рейтингах, а также небольшое количество пропусков в годах выпуска игры. Возможно эти пропуски возникли в результате человеческого фактора (упустили заполнение указанных данных), либо неправильного слияния таблиц (ошибка при объединении данных), либо такие данные отсутствуют в принципе.  

Далее мы посмотрим где можно найти адекатную замену пропускам и их обработаем, в противном случае некоторые пропуски придётся удалить.


```python
data[data['year_of_release'].isnull()]
```

Беглый поиск в открытых источниках даёт следующую пищу для размышлений: например, игра FIFA Soccer 2004 (строка 377) выходила с декабря 2003 по январь 2004 в разных странах. Для PDC World Championship Darts 2008 (строка 16373) первый релиз был с января 2008 по июнь 2009 для разных платформ и стран. А Hakuouki: Shinsengumi Kitan (строка 16458) вышла на	PS3	только в 2014 году, хотя первый релиз игры состоялся в 2008 году.  
Таким образом, пропуски в годах выхода игр могут быть связаны либо с разными годами релиза для разных стран, либо с выходами старой игры на новой платформе.  
В таком случае, адекватное заполнение данных пропусков не представляется приемлемым, учитывая соотношение размера пропусков к общему объёму данных (2%), решено данные пропуски удалить.  
С двумя пропусками в столбце name и genre поступим также, ввиду их незначительности.

```python
data = data.dropna(subset=['name'])
data = data.dropna(subset=['genre']).reset_index(drop=True)
data = data.dropna(subset=['year_of_release']).reset_index(drop=True)
```

 Проанализировав строки с пропусками и состав данных, сделали следующие предположения: пропуски в столбце user_score и значения "tbd" (to be determined) несут в себе одну смысловую информацию (оценка не определена).  


 Пропуски в графе rating могут быть связаны не только с ошибкой в данных, но и с тем, что для данных игр по каким-либо причинам оценка рейтинга ESRB не требовалась, либо при оценке игрового контента однозначно присвоить подходящую возрастную категорию не удалось.  


 Вообще, ESRB - это саморегулируемая организация, которая присваивает рейтинги по возрасту и контенту потребительским видеоиграм в Канаде, Соединенных Штатах и Мексике. В нашем датасете пристутствуют игры как из Северной Америки, так и Европы и Японии. Соответсвенно для игр, выпускаемых в этих странах, маркировки рейтинга ESRB быть не может, либо она не обязательна.


 Пропуски в столбце user_score и critic_score  заменить невозможно, т.к. все оценки индивидуальны. Аббревиатуру "tbd" заменим на NaN, т.к. она означает недостаточное количество оценок для выставления общей.  


 Заполним пропуски в столбце rating на значение 'undef'.
 Логичную замену в осталных столбцах предложить невозможно, поэтому на данном этапе оставим эти значения пустыми, а затем посмотрим на оставшееся количество пропусков.

```python
data['user_score']=data['user_score'].replace({'tbd':np.NaN})
data.rating = data.rating.fillna('undef')
data.isna().sum()
```

**Поиск и удаление полных дубликатов**

```python
# поищем явные дубликаты
data.duplicated().sum()
```

Полных соответствий не найдено, но возможно имеются неявные дубликаты, например игр, вышедших на одной и той же платформе в один год.

```python
data = data.drop_duplicates()

print(data[data.duplicated(['name', 'platform', 'year_of_release'])].count())
print(data[data.duplicated(['name', 'platform', 'year_of_release'])])
data = data.drop_duplicates(['name', 'platform', 'year_of_release'])
print(data[data.duplicated(['name', 'platform', 'year_of_release'])].count())
```

Были найдены и удалены дупликаты для двух игр.  
Теперь всё готово к следующему этапу - конвертированию типов данных.

### Преобразование типов данных

```python
data.info()
```

Очевидно, что данные в столбце year_of_release должны иметь тип int64, т.к. это всегда целые числа.  
Данные в столбце critic_score это тоже всегда целые числа, но, т.к. в них имеется большое количество пропусков,которые нельзя заменить логическими значениями и удалить также нельзя, - то на данном этапе оставим их float64.  
Данные столбца user_score - это в основном вещественные числа, которые также содержат NaN. Мы можем изменить тип данных на float64, при этом значения "tbd", если они остались, также станут NaN, на данном этапе это приемлемо.

```python
data['year_of_release'] = (data['year_of_release']).astype('int')
data['user_score'] = (data['user_score']).astype('float')
data.info()
```

### Посчитаем и добавим в таблицу данные о суммарных продажах во всех регионах

```python
# создадим новый столбец "total_sales" для общей суммы продаж игры
data['total_sales'] = data['na_sales'] + data['eu_sales'] + data['jp_sales'] + data['other_sales']
data.head()
```

**Вывод:** на данном этапе мы произвели предобработку данных.  
Исправлены названия столбцов, приведены к нижнему регистру.  
Дубликатов не обнаружено, удалены 2 строки не поддающиеся восстановлению, а так же 2% данных которые не подходят для исследования по причине отсутствия года выпуска.  
Заменены типа данных для столбцов необходимых для дальнейшей работы.  
Аббревиатура "tbd" заменена на NaN, т.к. она означает, недостаточное количество оценок для выставления общей.  

Удалены строки с неявными дубликатами поиск одновременно по 'name', 'platform', 'year_of_release'.
По столбцу name удалены 2 строки с пропусками.
По столбцу year_of_release принято решение удалить строки с пропусками.
В столбце critic_score количество пропусков больше 51% они оставлены как есть.
В столбце user_score количество пропусков больше 40%, неописанное значение "tbd" заменили на NaN.
В столбце rating пропуски обозначены как undef.
Столбцы year_of_release, critic_score приведены к int типу, user_score к float в соответствии с данными содержащимися в них.
Пропуски могли появиться из-за сбора данных из различных источников, которые не согласованы между собой, либо таких данных не сущестсвует, либо они противоречивы.
Посчитали суммарные продажи во всех регионах и записали их в total_sales.

## Исследовательский анализ данных

### Cколько было выпущено игр по годам


Составление таблицы с количеством выпускаемых игр по каждому году.

```python
data_years = pd.pivot_table(data, index='year_of_release', values='name', aggfunc='count')
data_years.columns = ['games']
data_years
```

Можем убедиться, что в датасете присутствуют данные по играм за 36 лет - с 1980 по 2016 годы.

```python
(
data
    .pivot_table(index='year_of_release', values='name', aggfunc='count')
    .plot(kind='bar', grid=True, figsize=(10, 5))
)
plt.title('Количество продаж по годам')
plt.xlabel('Год')
plt.ylabel('Количество продаж')
plt.show()
```

**Вывод:** как можно увидеть, относительно малое количество игр выпускалось до 1995 года, далее идет резкий рост с пиком в 2008-2009 годах. В 2010 году выпуск игр пошёл на спад и с 2012 года объем выпуска игр стабилизировался и год от года отличается не очень сильно.
На мой взгляд, данные до 1995 включительно можно не брать во внимание, а в целях исследования логично выбрать период в 5 лет (консоли быстро устаревают, а технические требования к играм возрастают).

### Исследование продаж по платформам


 Посмотрим, как менялись продажи игр в зависимости от платформы, на которой они выходили. Для начала определим платформы с наибольшими продажами по годам. Для этого построим график на снове сводной таблицы, распределим данные по убыванию чтобы сразу определить лучшие платформы.

### Влияние отзывов на продажи

Как мы уже выяснили, на текущий момент самой популярной  и перспективной консолью является PS4. Построим для неё диаграмму рассеяния и подсчитаем корреляцию между отзывами и продажами.

```python
# выберем для анализа данные за актуальный период и без пропусков в оценках пользователей
sns.scatterplot(
    x='user_score',
    y='total_sales',
    data=data_actual.query('platform == "PS4"').dropna(subset=['user_score']))
plt.title('Корреляция между отзывами пользователей и продажами PS4')
plt.xlabel('Оценка пользователей')
plt.ylabel('Суммарные продажи')
plt.show()
```

```python
# для оценки корреляции рассчитаем коэффициент Пирсона
print(f'Корреляция между оценками пользователей и продажами по платформе равна:',
      data_actual
      .query('platform == "PS4"')['user_score']
      .corr(data_actual.query('platform == "PS4"')['total_sales']))
```

Как мы видим на графике и исходя из полученного значения коррелляции - оценка пользователей слабо влияет на продажи игр на консоли PS4: продажи около 4 и 8 млн копий имеют достаточно среднюю оценку (обратный эффект популярности - чем более массовый продукт, тем больше у него имеется недовольных пользователей). А самые высокие оценки имеют как и не очень продаваемые игры, так и достаточно популярные.

```python
# выберем для анализа данные за актуальный период и без пропусков в оценках критиков
sns.scatterplot(
    x='critic_score',
    y='total_sales',
    data=data_actual
    .query('platform == "PS4"')
    .dropna(subset=['critic_score']))
plt.title('Корреляция между отзывами кртиков и продажами PS4')
plt.xlabel('Оценка критиков')
plt.ylabel('Суммарные продажи')
plt.show()
```

```python
# для оценки корреляции рассчитаем коэффициент Пирсона
print(f'Корреляция между оценками критиков и продажами по платформе PS4 равна:',
      data_actual
      .query('platform == "PS4"')['critic_score']
      .corr(data_actual.query('platform == "PS4"')['total_sales']))
```

Диаграмма распределения явно показывает на умеренную связь между оценкой критиков и продажами игр на PS4, соответствующий коэффициент корреляции подтверждает это.

## Проверка гипотез

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Сформируем нулевую гипотезу (Н0) из условия равенства что средние оценки пользователей игр user_score платформы XOne равны средним оценкам пользователей игр платформы PC.

Сформируем альтернативную гипотезу (Н1), которая проверяется, если отбрасывается Н0: средние оценки пользователей игр user_score платформы XOne не равны средним оценкам пользователей игр платформы PC.

Т.к. нам требуется проверить гипотезу о равенстве среднего двух независимых выборок, то для её проверки воспользуемся методом ttest_ind с параметром equal_var=False, т.е. есть основания предполагать, что их дисперсии разные.   
Зададим пороговое значение уровня значимости alpha = 0.05. Выберем для анализа данные без пропусков в оценках пользователей.

Итак, выдвинем гипотезы:
* Н0: Средние пользовательские рейтинги платформ XOne и PC равны
* Н1: Средние пользовательские рейтинги платформ XOne и PC не равны

```python
# выберем для анализа актуальный период и удалим пропуски в данных
hypothesis = (data_actual.dropna(subset=['user_score']))

# формулируем t-test, сравниваем выборки с оценками пользователей по платформам
results = st.ttest_ind(hypothesis[hypothesis.platform =='XOne']['user_score'],\
                       hypothesis[hypothesis.platform =='PC']['user_score'],\
                       equal_var=False, alternative='two-sided')
alpha = 0.05
# выводим значения p-value на экран
print(results.pvalue)
# условный оператор с выводом строки с ответом
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Не получилось отвергнуть нулевую гипотезу, есть основания утверждать, что рейтинги платформ XOne и PC равны, при установленном уровне значимости в 5 %.

### Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные

Будем исходить от противного и сформируем нулевую гипотезу (Н0) из условия равенства, что средние оценки пользователей игр user_score жанров Action равны средним оценкам пользователей игр жанров Sports.

Сформируем альтернативную гипотезу (Н1), которая проверяется, если отбрасывается Н0: средние оценки пользователей игр user_score жанров Action не равны средним оценкам пользователей игр жанров Sports.

Для проверки гипотезы также будем использовать метод ttest_ind с параметром equal_var=False, т.к. есть основания полагать, что дисперсии этих двух выборок разные. Зададим пороговое значение уровня значимости alpha = 0.05.

Выдвинем гипотезы:

* Н0: Средние пользовательские рейтинги жанров Action и Sports равны
* Н1: Средние пользовательские рейтинги жанров Action и Sports не равны

```python
# формулируем t-test, сравниваем выборки с оценками пользователей по жанрам
results = st.ttest_ind(hypothesis[hypothesis.genre =='Action']['user_score'],\
                       hypothesis[hypothesis.genre =='Sports']['user_score'],\
                       equal_var=False, alternative='two-sided')
alpha = 0.05
# выводим значения p-value на экран
print(results.pvalue)
# условный оператор с выводом строки с ответом
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Нулевая гипотеза отвергнута и исходя от противного, есть основания утверждать, что рейтинги жанров Action и Sports не равны, а значит разные, при установленном уровне значимости в 5 %.

**Пояснения:**

Альтернативная гипотеза может быть проверена в трёх направлениях "меньше", "не равно", "больше". В нашем случае ставится вопрос про равенство, значит проверку надо провести двухстороннюю: alternative='two-sided'.

По заданию проверить, что "Средние пользовательские рейтинги жанров Action и Sports разные" нельзя проверить прямо разность выборок, ведь нулевая гипотеза должна наоборот подтверждать равность выборок, поэтому нами была выдвинута следующая нулевая гипотеза Н0: Средние пользовательские рейтинги жанров Action и Sports равны. На 2016 год нет явно возрастающих продаж по консолям, продажи падают везде: пользователи видимо переходят в сегмент мобильных игр или играют на домашнем компьютере, который проще и дешевле обновить. Более или менее перспективными для западных стран являются консоли PS4 и XOne, но у них также прослеживается тенденция к падению продаж.
Установлено, что оценки пользователей и критиков не сильно влияют на продажи, кассовые игры имеют нередко низкий рейтинг.
Может быть, веянее времени, но предпочтения в жанрах сейчас это стрелялки и спортивные симуляторы, видимо те игры, в которые можно играть время от времени и не увлекаться прогрессом в игре.  

Портрет пользователя компьютерных игр оказался таким: типичный пользователь игр в Северной Америке приобретает игры платформ PS4 и XOne жанров Action, Shooter или Sports из возрастных категорий M («Для взрослых», 17+) и E («Для всех», 6+) или без рейтинга ESRB.   
Типичный пользователь игр в Европе приобретает игры платформ PS4 и XOne жанров Action, Shooter или Sports из возрастных категорий M («Для взрослых», 17+) и E («Для всех», 6+) или без рейтинга ESRB.   
Типичный пользователь игр в Японии приобретает игры платформ 3DS и PS4 жанра Role-Playing и Action, не имеющей рейтинга возрастных категорий ESRB. В Японии игры платформы Xone (одного из лидеров рынка Северной Америки и Европы) практически не продаются, игры жанра Sports (одного из лидеров рынков Северной Америки и Европы) - продаются очень мало. Абсолютно большая часть продаваемых игр в Японии не имеет рейтинга ESRB.

Средние пользовательские рейтинги для платформ Xbox и PC различны, и пользовательские рейтинги жанров Action и Sports тоже не одинаковы.  

**Рекомендации:**
Считаем, что при прогнозировании рекламной компании и маркетинговой стратегии надо опираться на следующие факты:  делать ставку на консоли нового поколения в зависимости от популярных консолей в регионе (необходимо иметь ввиду, что рынок Японии не похож на рынок западных стран и учитывать это в работе). Такая же ситуация с жанром и возрастному рейтингу: судя по полуярности рейтинга M - основная часть геймеров старше 17 лет. Также стоит учитывать рейтинг критиков на вышедшую игру, чем он выше тем, скорее всего больше, будет продаж.
Стоит учесть, что в конце актуального периода продажи игра падали по всем платформам, вероятно выручка в 2017 году будет меньше, чем в 2016, пока не появятся консоли нового поколения, т.к. после выхода новых консолей резко начинают расти продажи игр.
Можно отметить, что стиабильными являются продажи игр на платформе PC.
В целях увеличения прибыли интернет магазина "Стримчик" необходимо запланировать проведение рекламных компаний игр на потенциально популярные продукты, а именно: платформа PS4, ХBox, жанры Shooter и Sports с рейтингом "M" (для взрослых).
